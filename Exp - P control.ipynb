```python
import numpy as np
import matplotlib.pyplot as plt
from differential_robot import DifferentialRobot
```

```python
sample_time = 0.01 # [s]
simulation_time = 100 # [s]
simulation_steps = int(simulation_time / sample_time)
```

```python
P = 2

robot = DifferentialRobot(0, 0.1, 0.1, sample_time, 0.025, 0.1, 1.2, 0.001)
robot.step_wheel(1, 1)
leituras = []
for _ in range(simulation_steps):
    if _ == 600: #_ == 600 or
        robot.move(1)
    sensor = robot.line_sensor()
    leituras.append(sensor)
    a_left = 1 + sensor*P
    a_right = 1 - sensor*P
    robot.d_speed(a_left, a_right)
plt.plot(leituras)
plt.plot([0, simulation_steps], [0.0, 0.0], '--')
plt.show()
```

```python


for y in np.linspace(-0.3, 0.3, 6):
    for P in np.linspace(0.2, 2.5, 6):
        robot = DifferentialRobot(0, y, 0.2, sample_time, 0.025, 0.1, 1.2, 1)
        leituras = []
        for _ in range(simulation_steps):
            sensor = robot.line_sensor()
            leituras.append(sensor)
            a_left = 4 + sensor*P
            a_right = 4 - sensor*P
            robot.step_acc(a_left, a_right)
        plt.plot(leituras)
        print(f'{robot.state()=}')
    plt.plot([0, simulation_steps], [0.0, 0.0], '--')
    plt.show()
```

```python
P = 4

robot = DifferentialRobot(0, 0, 0.1, sample_time, 0.025, 0.1, 1.6, 0.0004)
robot.step_wheel(1, 1)
leituras = []
for _ in range(simulation_steps):
    if _ == 1000:
        robot.move(-0.9)
    sensor = robot.line_sensor()
    leituras.append(sensor)
    a_left = 1 + sensor*P
    a_right = 1 - sensor*P
    robot.d_speed(a_left, a_right)
plt.plot(leituras)
plt.plot([0, simulation_steps], [0.0, 0.0], '--')
plt.show()
```

```python
robot = DifferentialRobot(0, 0.1, 0, sample_time, 0.025, 0.1, 1.6, 0.0004)
leituras = []
f = True
for _ in range(simulation_steps):
    if robot.sensor_x() >= 0.4828427 and f:
        robot.move(-np.pi/2)
        f = False
    sensor = robot.line_sensor()
    leituras.append(sensor)
    robot.step_wheel(1, 0.9)
plt.plot(leituras)
plt.plot([0, simulation_steps], [0.0, 0.0], '--')
print(f'{robot.state()=}')
plt.show()
```
